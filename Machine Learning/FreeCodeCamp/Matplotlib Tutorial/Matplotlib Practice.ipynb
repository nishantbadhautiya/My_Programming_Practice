```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
labels = ['Banana', "Mango", "Apple", "Orange"]
values = [23,34,56,78]
plt.figure(dpi=100, figsize=(8,5))
bars = plt.bar(labels, values)
patterns = ["*", "o", "/","."]

for bar in bars:
    bar.set_hatch(patterns.pop())

plt.yticks([20,40,60,80])
plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")
gas
```

```python
font = {
    "fontname" : "serif",
    "fontsize" : 12,
    "fontweight" : "bold",
    "color" : "green"
}
plt.figure(dpi=200, figsize=(8,5))
plt.title("Gas Price in difference country per gallon", fontdict=font)
plt.xlabel("Year", fontdict=font)
plt.ylabel("Country", fontdict= font)
countries_to_look_at = ["Australia","Canada", "France", "Germany"]
for country in gas:
    # if country != "Year":
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], 'o-', ms=3, mec='r', mfc = 'r', label = country)

plt.xticks(gas.Year[::3].to_list() + [2011])
plt.legend()
plt.show()
```

```python
fifa = pd.read_csv("fifa_data.csv")
fifa
```

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa["Overall"], bins=bins, color='#611630')
plt.xlabel("Overall Score")
plt.ylabel("No. pf Players")
plt.title("Distribution of Players skill in FIFA 2018")
plt.show()

```

```python
left = fifa.loc[(fifa['Preferred Foot'] == 'Left')].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
labels = ['Left', 'Right']
colors = ['#aabbcc', '#aa00bb']
hatches = ["*", "o"]
plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f%%', shadow=True, radius=1.2)
plt.title("Foot Preference of FIFA Players")
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight > 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight > 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight > 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight > 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Above 200']
explodes = [0.4, 0, 0, 0, 0.4]

plt.pie(weights, labels= labels, autopct='%.2f %%', pctdistance=0.6, explode= explodes)
plt.title("Weight Distribution of different Players in FIFA 2018 (in lbs)", loc='center')
plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(8,12))
fc_barcelona = fifa.loc[(fifa['Club'] == "FC Barcelona")]['Overall']
tranmere_rovers = fifa.loc[(fifa['Club'] == "Tranmere Rovers")]['Overall']
manchester_united = fifa.loc[fifa['Club'] == "Manchester United"]['Overall']
labels = ['fc_barcelona', 'tranmere_rovers', 'manchester_united']

bp = plt.boxplot([fc_barcelona, tranmere_rovers, manchester_united], labels= labels, patch_artist=True)
for box in bp['boxes']:
    # set the color
    box.set(color = "r")
    # change fill color
    box.set(facecolor = 'g')

plt.title("Game Club Performance in FIFA")
plt.ylabel("Overall Score")
plt.show()
```
